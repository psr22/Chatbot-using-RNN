```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from nltk.corpus import wordnet as wn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes, svm
from sklearn.metrics import accuracy_score

```

```python
np.random.seed(500)
```

```python
Corpus = pd.read_csv(r"corpus.csv",encoding='latin-1')
```

```python
Corpus
```

```python
Corpus['text'].dropna(inplace=True)
```

```python
Corpus['text'] = [entry.lower() for entry in Corpus['text']]
```

```python
Corpus['text']= [word_tokenize(entry) for entry in Corpus['text']]
```

```python
Corpus.head
```

```python
tag_map = defaultdict(lambda : wn.NOUN)
tag_map['J'] = wn.ADJ
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV
```

```python
print(tag_map)
word_Lemmatized = WordNetLemmatizer()
Corpus['text']
```

```python
for index, entry in enumerate(Corpus['text']):
    # print(index, entry)
    # print("--------------------------------")
    Final_words = []
    for word, tag in pos_tag(entry):
        # print(word, tag)
        if word not in stopwords.words('english') and word.isalpha():
            word_Final = word_Lemmatized.lemmatize(word,pos = tag_map[tag[0]])
            Final_words.append(word_Final)
    Corpus.loc[index,'text_final'] = str(Final_words)
```

```python
Corpus
```

```python
Train_X, Test_X, Train_Y, Test_Y = model_selection.train_test_split(Corpus['text_final'],Corpus['label'],test_size=0.3)
```

```python
Encoder = LabelEncoder()
Train_Y = Encoder.fit_transform(Train_Y)
Test_Y = Encoder.fit_transform(Test_Y)
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(Corpus['text_final'])
```

```python
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
NB = naive_bayes.MultinomialNB()
NB.fit(Train_X_Tfidf,Train_Y)

# predict the labels on validation dataset
pred_NB = NB.predict(Test_X_Tfidf)

# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(pred_NB, Test_Y)*100)
```

```python
SVM = svm.SVC(C = 1.0, kernel = 'linear', degree = 3, gamma = 'auto')
```

```python
SVM.fit(Train_X_Tfidf, Train_Y)
```

```python
pred_SVM = SVM.predict(Test_X_Tfidf)
```

```python
print("SVM Accuracy Score -> ",accuracy_score(pred_SVM, Test_Y)*100)
```
